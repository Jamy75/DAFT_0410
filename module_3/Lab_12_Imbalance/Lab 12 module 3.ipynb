#### Check the datatypes of all the columns in the data.You would see that the column TotalCharges is object type.

Convert this column into numeric type using pd.to_numeric function.

```python
change = ['TotalCharges']
costumer[change] = costumer[change].apply(pd.to_numeric,errors='coerce')
```

```python
costumer.info()
```

#### Check for null values in the dataframe. Replace the null values.

```python
missing_values = costumer[costumer.isnull().any(axis=1)]
missing_values
```

```python
costumer["TotalCharges"].fillna((costumer['TotalCharges'].mean()), inplace = True)
costumer
```

### Use the following features: tenure, SeniorCitizen, MonthlyCharges and TotalCharges:

- Scale the features either by using MinMaxScaler or a standard scaler.

```python
X = costumer[['tenure', 'SeniorCitizen', 'MonthlyCharges', 'TotalCharges']]
#y = costumer['tenure', 'SeniorCitizen', 'MonthlyCharges', 'TotalCharges']
y = costumer['Churn']
```

```python
from sklearn.model_selection import train_test_split
```
